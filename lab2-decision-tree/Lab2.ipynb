```python
import pandas as pd
df = pd.read_csv('lab2.csv')
df.sample(10)
```

```python
from sklearn.model_selection import train_test_split
# Splitting data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    df.drop('label', axis=1), df['label'],
    test_size=0.25, random_state=17)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize classifier model
clf = DecisionTreeClassifier(criterion='entropy',max_depth=4, 
                             min_samples_split=5, min_samples_leaf=3, min_weight_fraction_leaf = 0.0, 
                             max_features = None, random_state=None, max_leaf_nodes=5, 
                             min_impurity_decrease=0.0, min_impurity_split=None, class_weight=None, ccp_alpha=0.0);

# Train the classifier model on training data (model fitting)
clf.fit(X_train, y_train);
```

```python
# Make predictions on test data
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
# Compare test set predictions with ground truth labels
print("Accuracy = %.3f" % (accuracy_score(y_pred, y_test)))
```

```python
f_names = list(df.columns)
len(f_names)
```

```python
f_names.remove('label')
len(f_names)
```

```python
importance = clf.feature_importances_
print("Feature importances:")
for i,v in enumerate(importance):
    print('Feature: %s, Score: %.5f' % (f_names[i],v))
```

```python
f_scores = pd.DataFrame({'feature':f_names,
                    'score': importance})
```
